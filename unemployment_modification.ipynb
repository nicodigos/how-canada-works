# Unemployment Data

```python
df = pd.read_csv("Unemployment_Canada_1976_present.csv")
```

```python
df.GEO.unique()
```

```python
df_Canada = df[~(df.GEO == 'Canada')]
```

```python
df_Canada_2 = df_Canada[df_Canada['Age group'].isin(['15 years and over'])]
```

```python
# Check if we have '15 to 24 years', '25 years and over' for all months available and all provinces
comparison_set = {'15 years and over'}
notifications = []
for province in df_Canada_2.GEO.unique():
    temp_df = df_Canada_2[df_Canada_2.GEO == province]
    for month in temp_df['REF_DATE'].unique():
        if set(temp_df[temp_df.REF_DATE == month]['Age group'].tolist()) == comparison_set:
            continue
        else:
            notifications.append(f'{province}, {month}')
```

```python
notifications
```

```python
notifications = []
for province in df_Canada_2.GEO.unique():
    temp_df = df_Canada_2[df_Canada_2.GEO == province]
    notifications.append(f'{province}, months_available: {len(temp_df["REF_DATE"].unique())}')
```

```python
notifications
```

```python
df_Canada_2['year'] = df_Canada_2.REF_DATE.str[:4]
```

**First chart for unemployment**

```python
df_province_year_employment = df_Canada_2[['GEO', 'year', 'Labour force', 'Employment', 'Population', 'Part-time employment ', 'Full-time employment']].groupby(['GEO', 'year']).mean().reset_index()
```

# Total Population Data

```python
population = pd.read_csv('17100005-eng/17100005.csv')
population_metadata = pd.read_csv('17100005-eng/17100005_MetaData.csv')
```

```python
pop_total_gender = population[population.Gender == 'Total - gender']
```

```python
pop_all_ages = pop_total_gender[pop_total_gender['Age group'] == 'All ages']
pop_15_less = pop_total_gender[pop_total_gender['Age group'].isin(['0 to 4 years', '10 to 14 years'])] 
```

```python
pop_only_provinces = pop_all_ages[pop_all_ages.GEO.isin(df_Canada.GEO.unique().tolist())]
pop_15_only_provinces = pop_15_less[pop_15_less.GEO.isin(df_Canada.GEO.unique().tolist())]
```

```python
pop_only_provinces.REF_DATE = pop_only_provinces.REF_DATE.astype(str)
pop_15_only_provinces.REF_DATE = pop_15_only_provinces.REF_DATE.astype(str) 
```

```python
result_1 = pd.merge(df_province_year_employment, 
         pop_only_provinces[['REF_DATE', 'GEO', 'VALUE']],
        how='left', 
         left_on=['GEO', 'year'], 
         right_on=['GEO', 'REF_DATE'])
```

```python
result_1
```

```python
pop_15_only_provinces_grouped = pop_15_only_provinces[['GEO', 'REF_DATE', 'VALUE']].groupby(['GEO', 'REF_DATE']).sum().reset_index()
```

```python
result_2 = pd.merge(result_1, 
                    pop_15_only_provinces_grouped, 
                    how='left', 
                    on = ['GEO', 'REF_DATE']
                   ).rename(columns={'VALUE_x': 'Total_Population', 'VALUE_y':'Population_0_14'})
```

```python
result_2
```

```python
result_2['Population_15_more'] = result_2.Total_Population - result_2.Population_0_14
result_2['per_labour_force'] = result_2['Labour force'] / result_2.Population
result_2['per_employment'] = result_2.Employment / result_2.Population
result_2['per_unemployment'] = result_2['per_labour_force'] - result_2['per_employment']
result_2['per_parttime'] = (result_2.Employment * result_2['Part-time employment ']/
                            (result_2['Part-time employment '] + result_2['Full-time employment'])) / result_2.Population
result_2['per_fulltime'] = (result_2.Employment * result_2['Full-time employment']/
                            (result_2['Part-time employment '] + result_2['Full-time employment'])) / result_2.Population
```

```python
result_2['labour_force'] = result_2['Population_15_more'] * result_2['per_labour_force']
result_2['inactive_adults'] = result_2['Population_15_more'] - result_2['labour_force']
result_2['employment'] = result_2['Population_15_more'] * result_2['per_employment']
result_2['unemployment'] = result_2['Population_15_more'] * result_2['per_unemployment']
result_2['parttime'] = result_2['Population_15_more'] * result_2['per_parttime']
result_2['fulltime'] = result_2['Population_15_more'] * result_2['per_fulltime']
```

```python
result_2.columns
```

**First Chart for Unemployment**

```python
result_2[['GEO', 'year',
       'Total_Population', 'Population_0_14', 'Population_15_more', 'labour_force', 'inactive_adults', 'employment',
       'unemployment', 'parttime', 'fulltime']].to_csv("data_unemployment_first_chart.csv", index=False)
```
